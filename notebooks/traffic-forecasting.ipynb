<h1><b> Define Functions

```python
def all_weeks(a):
    w_1= a.iloc[:168,:] 
    w_2=a.iloc[168:336,:]
    w_3= a.iloc[336:504,:]
    w_4=a.iloc[504:672,:]
    w_5= a.iloc[672:840,:]
    return w_1,w_2,w_3,w_4,w_5
```

```python
def melt_(a_,b_,c_,d_,e_):
    a=a_.melt(var_name='column', value_name='value')
    b=b_.melt(var_name='column', value_name='value')
    c=c_.melt(var_name='column', value_name='value')
    d=d_.melt(var_name='column', value_name='value')
    e=e_.melt(var_name='column', value_name='value')
    repeating_index = np.tile(np.arange(168), int(np.ceil(483840 / 168)))[:483840]
    a['index'] = repeating_index
    b['index'] = repeating_index
    c['index'] = repeating_index
    d['index'] = repeating_index
    e['index'] = repeating_index
    return a,b,c,d,e
```

```python
def date_(n):
    hours_per_day = 24
    hours_per_week=168
    days_per_month = 30
    hours_per_month = hours_per_day * days_per_month

    n['week'] = (n['index'] ) // hours_per_week + 1
#     n['day'] = ((n['index'] ) % hours_per_week) // hours_per_day  + 1
    n['month'] = np.abs(a1['index'] - 1) // hours_per_month + 1
    n['day_of_month'] = (np.abs(a1['index'] - 1) % hours_per_month) // hours_per_day + 1
#     n['day_of_week'] = (np.abs(a1['index'] - 1) % (hours_per_day * 7)) // hours_per_day + 1
    n['hour'] = n['index'] % 24
    n['day_of_week'] = np.abs(n['index'] // 24) % 7
    n['day'] = n['index'] // 24
    return n
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import SGDRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import BaggingRegressor
from catboost import CatBoostRegressor
from lightgbm import LGBMRegressor
from xgboost import XGBRegressor
from sklearn.model_selection import TimeSeriesSplit
X=train.drop("Lag",axis=1).values
y=train["Lag"].values
scores=[]
kf=TimeSeriesSplit(n_splits=5)
for train_index,test_index in kf.split(X,y):
    X_train,y_train=X[train_index],y[train_index]
    X_test,y_test=X[test_index],y[test_index]

    xg= RandomForestRegressor(random_state=42)
    xg.fit(X_train,y_train)
    y_pred=xg.predict(X_test)
    scores.append(mean_absolute_error(y_pred,y_test))

print("mean:",np.mean(scores))
```
